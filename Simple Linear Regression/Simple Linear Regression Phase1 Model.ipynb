```python
#SIMPLE LINEAR REGRESSION MODEL CREATION AND EVALUATION

#Pandas is a powerful, open-source Python library used for data manipulation and analysis. 
#It is a data preparation powerhouse. 
#You'll use it to load, clean, explore, and prepare your data before feeding it to sklearn for machine learning
import pandas as pd

#importing data set to 'Dataset' variable using pandas and read_csv function
Dataset = pd.read_csv("Salary_Data.csv") #loading dataset
Dataset

#declaring input and output variables from the dataset for analysis
Independent = Dataset[["YearsExperience"]]
Independent
Dependent = Dataset[["Salary"]]
Dependent

#Spliting train and test data from dataset to evaluate and analyse 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Independent, Dependent, test_size = 0.30, random_state = 0) 

#bulding Simple linear regression model
#Sklearn -> Scikit-learn is a most popular machine learning library in Python
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression()
# fit(x,y) -> It is an SKLearn API where it trains the model and is always used first
regressor. fit(x_train, y_train)

#calculating weight and bias
weight = regressor.coef_
bias = regressor.intercept_
print(f" Bias for the linear model from the data set is {bias}")
print(f" Weight for the linear model from the data set is {weight}")

#Y-prediction -> predicting the SRL from test data set 
y_predict = regressor.predict(x_test)

#Evaluation metrics using r2 using predicted value and test data's actual value
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_predict)
print(f" The r2 score for the linear regression model is {r_score}")
if r_score > 0 and r_score < 1 :
    print(f"The predicted Simple Linear Regression Model has better performance with {r_score} r2 score")
else:
    print(f"The predicted Simple Linear Regression Model is not having better performance with {r_score} r2 score")

#Save the model
import pickle
#It saves the trained machine learning models to your hard drive so you can use them later without retraining!

filename = "Final_model_SLR.sav" #declaring a pickle file name
pickle.dump(regressor,open (filename, 'wb'))#creating a pickle file in the local repo
loaded_model = pickle.load(open("Final_model_SLR.sav", 'rb')) #loading the pickle file 
result = loaded_model.predict([[12]])
print(f'The predicted salary that can be given for the given years of experiance is {result}')
```
