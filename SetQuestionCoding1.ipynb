```python
'''4.Write a Python function to find the factorial of 5. Expected Output:
120'''

'''The logic is:

1. Initialize factorial = 1
2. Loop from 1 to the number
3. Multiply factorial by each number (factorial *= i)
4. Print the final result once after the loop completes'''
```

```python
Factorial = 1
num = 5

for i in range(1, num+1):
    Factorial *= i
print(Factorial)
```

```python
'''5.Write a Python program to check if a number is prime. Expected Output (for input 7):
7 is a prime number'''

'''
Step 1: If number <= 1 → NOT PRIME (return False)

Step 2: Loop i from 2 to (number - 1)
        Check if number is divisible by i (number % i == 0)
        If YES → NOT PRIME (return False)

Step 3: If loop completes without finding any divisor → PRIME (return True)'''
```

```python
num = 7
is_prime = True

if num <= 1:
    is_prime = False
else:
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
            
if is_prime:
    print(f"{num} is a prime number")
else:
    print(f"{num} is not a prime number")
```
