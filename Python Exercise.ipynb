```python
#BMI Calculator with Category Classification
```

```python
weight = float (input("Enter weight in Kg : "))
Height = float (input("Enter Height in Cm : "))
#convert cm to m
Height_in_m = Height /100
#calculate BMI
BMI = weight / Height_in_m
#Display calculations
print(" Your Weight in Kg is : ",weight)
print("Your Height in meters is : ", Height_in_m)
print("Your BMI is : ", BMI)
#Classifying the category
if BMI < 18.5:
    print("Status: Underweight")
elif BMI < 25:
    print("Status: Healthy weight")
elif BMI < 30:
    print("Status: Overweight")
else:
    print("Status: Obese")
```
